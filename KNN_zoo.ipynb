```python
acc=[]
k_i=[]
for i in  range(3,50,2):
    KNN_1=KNC(n_neighbors=i)
    KNN_1.fit(x_train,y_train)
    pred_1=KNN_1.predict(x_test)
    acc_1=np.mean(pred_1==y_test)
    k_i.append(i)
    acc.append(acc_1)
    accuracy=pd.DataFrame({'neigh':k_i,'accuracy':acc})
```

```python
accuracy
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(3,50,2),acc,linestyle='dashed',marker='o',markersize=10)
plt.xlabel('k_neighbours')
plt.ylabel('accuracy')
```
